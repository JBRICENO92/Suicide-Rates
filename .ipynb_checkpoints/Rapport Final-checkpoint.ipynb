Suicide is an extremely complex and sensitive tragedy that causes pain to hundreds of thousands of people every year. The World Health Organisation (WHO) and the Global Burden of Disease study estimate that almost 800,000 people commit suicide every year. That's roughly two persons per second.

According to the World Health Organisation, social, psychological, cultural and many other factors contribute to  increase the risk of suicidal behaviour, but the stigma attached to suicide means that many people who are in need of help feel unable to seek it. The risk factors include but are not limited to previous suicide attempts, mental health problems, abusive use of alcohol, drug use, job or financial loss, relationship breakdowns, trauma or abuse, violence, conflict or disaster, and chronic pain or illness (WHO, 2014).

## **1.1. Suicide Rates And Healthy Goverment policies** <a class="anchor" id="interest"></a>



Suicide is a global burden that no country is exempt from. Countrues are in a constant race against time to reduce the number of fatalities. Political, economical and social factors ammong others influence the numbers of suicides committed. Goverments aroud the world would have tried to reduce the death toll if they had an tool/oracle to predict the numbers of suicide for each year. Subsequently, Goverment officials could determine the optimal public policies thus minimising the loss of human life.

The World Health Organization(WHO) recognizes suicide as a public health priority. In its first Suicide Report : “Preventing suicide: a global imperative”, published in 2014, WHO aimed to increase the awareness towards the public health significance of suicide and suicide attempts and to make suicide prevention a high priority on the global public health agenda. The report also sought to incite goverments to take action and develop or strengthen existing comprehensive suicide prevention strategies in a multisectoral public health approach. Some oossible strategies  are to : collect suicide data (number of casualties, estimate the rates, ...) and/or find informative variables.
Moreover, death by suicide is considered as the indicator of target 3.4 (target 3 sub-target 4) of the sustainable Development goals : "Reduce by one third premature mortality from noncommunicable diseases through prevention and treatment, and promote mental health and well-being by 2030". Hence, measuring suicide rates is a metric of the effiency of preventive and curative policies. Unfortunately, Data collection still poses a major hurdle in many countries either due to the lack of the appropriate infrastructure or to any other reason (change of the political regime, policies, ...).

Suicide is no longer considered a crime in developed countries but continues to be so in many developing countries, for example in islamic countries suicide is forbidden due to religious reasons.  Social factors also affect the censorship of suicide figures. some families do not disclose the death cause for fear of harassement and/or shame. Instead they claim it to be either an accident or in some cases, a homicide/murder. 

One can notice also that in developing countries, data acquisition and collection and estimating suicide rates don't score high on the priorities level. These countries usaully suffer friom more pressing matters such as poverty allevation, low literacy rates and high population growth. Besides, communicable diseases, maternal and child health problems  take precedence over non-communicable diseases. Mental health has very low priority in these situations.

##### Utility functions

````python
def plot_nan(data, var_cols, graphic = True, verbose = False):
    if graphic :
        d = {}
        for var in var_cols:
            d[var] = data[var].isna().sum()
        df = pd.DataFrame(d.items(), columns=['Name', 'Missing value count'])
        fig = px.pie(df, values='Missing value count', names='Name', color_discrete_sequence=px.colors.sequential.RdBu)
        fig.show()
    if verbose:
        for var in var_cols:
            print(f'There are {data.shape[0]-data[var].isna().sum()} available instances out of {data.shape[0]} for variable {var}')
````

### **2.5. Suicide Rates** <a class="anchor" id="suiciderates"></a>

The total rate column corresponds to the Suicide Mortality Rates. Viewing suicide as a share of total death is a way of comparing differences around the world. However, this does not give us a direct comparison of suicides between countries. This proportion is influenced by the magnitude of other causes of death.

To obtain an isolated comparison of suicides worldwide we will use suicide rates, which measure the number of suicides per 100,000 people in a given population.

Globally, 10 people per 100,000 died by suicide in 2017. On the map we show the differences in suicide rates around the world.

Suicide rates vary from about 20 per 100,000 in Eastern Europe, South Korea, Zimbabwe, Guyana and Suriname to less than 5 per 100,000 in North Africa, the Middle East, Indonesia, Peru and some Mediterranean countries.

Source: [Our World in Data : Suicide](https://ourworldindata.org/suicide#:~:text=Summary,in%202017%20were%20from%20suicide.)


````python
var =  'rate-total' #Desired variable
df = data #the data to use
path = './data/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp' # geopandas data path, you shouldn't modify this


def country_data(path):
    shapefile = path
    #Read shapefile using Geopandas
    gdf = gpd.read_file(shapefile)[['ADMIN', 'ADM0_A3', 'geometry']]
    #Rename columns.
    gdf.columns = ['Country', 'Country_code', 'geometry']
    return gdf

def json_data(path, data, year, var):
    df = data[data['Year']==year][['Country Code', 'Year', var]]
    gdf = country_data(path)
    merged = gdf.merge(df, left_on = 'Country_code', right_on = 'Country Code', how = 'left')
    merged.fillna('No data', inplace = True)
    merged_json = json.loads(merged.to_json())
    json_data = json.dumps(merged_json)
    return json_data, df


min_year, year, max_year = df['Year'].min(), df['Year'].min(), df['Year'].max()

jdf, df_aux = json_data(path, df, year, var)

#Input GeoJSON source that contains features for plotting.
geosource = GeoJSONDataSource(geojson = jdf)

#Define a sequential multi-hue color palette.
palette = brewer['YlGnBu'][8]

#Reverse color order so that dark blue is highest obesity.
palette = palette[::-1]

#Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors. Input nan_color.
color_mapper = LinearColorMapper(palette = palette, low = df_aux[var].min(), high = df_aux[var].max(),
                                 nan_color = '#d9d9d9')
#Add hover tool
hover = HoverTool(tooltips = [ ('Country/region','@Country'),(var, '@{%s}'%var)])

#Create color bar. 
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8,width = 500, height = 20,
                     border_line_color=None,location = (0,0), orientation = 'horizontal',)

#Create figure object.
p = figure(title = 'World map showing %s in %d' % (var, year), plot_height = 600,
           plot_width = 950, toolbar_location = None, tools = [hover])
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

#Add patch renderer to figure. 
p.patches('xs','ys', source = geosource,fill_color = {'field' : var, 'transform' : color_mapper},
          line_color = 'black', line_width = 0.25, fill_alpha = 1)

#Specify layout
p.add_layout(color_bar, 'below')
# Define the callback function: update_plot
def update_plot(attr, old, new):
    selectedyear = slider.value
    new_data, _ = json_data(path, df, selectedyear, var)
    geosource.geojson = new_data
    p.title.text = 'World map showing %s in %d' % (var, selectedyear)

# Make a slider object: slider 
slider = Slider(title = 'Year', start = min_year, end = max_year, step = 1, value = min_year)
slider.on_change('value', update_plot)

# Make a column layout of widgetbox(slider) and plot, and add it to the current document
layout = column(p,Column(slider))
curdoc().add_root(layout)

#Display plot inline in Jupyter notebook
output_notebook()

# show(layout)
````

This analysis is one of the methods used to decide which characteristics most affect the target variable and, in turn, are used to predict this target variable. In other words, it is a commonly used method for selecting characteristics in machine learning.
And since visualization is generally easier to understand than reading tabular data, heatmaps are typically used to visualize correlation matrices.


For this reason we have decided to include this thermal correlation map in order to know which features are strongly correlated and which are not. This analysis will help us to know which features to choose for the preprocessing and the creation of our estimator.

In the thermal map above we can see that

* Which is the strongest correlation pair and which is the weakest (except for the main diagonal)?
* What are the three most correlated variables?

The size of each square corresponds to the magnitude of the correlation it represents, that is
size(c1, c2) ~ abs(corr(c1, c2))

### **3.1. Preprocessing <a class="anchor" id="prep"></a>**

#### Missing Values:

Since our dataset has many missing values, it is necessary to do a pre processing of our data, using Sklearn's methods. In this example we will use the ```SimpleImputer``` and ```KNNImputer```.

````python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

S_imp = SimpleImputer(missing_values=np.nan, strategy='mean')
K_imp = KNNImputer(missing_values=np.nan, n_neighbors=5, weights="uniform")
````

#### Feature extraction:

In our problem, there are a lot of attributes, we may use Principal Componenent Analysis, to reduce the number of less informative variables.

````python
from sklearn.decomposition import PCA
pca = PCA(n_components = 10)
````

### **3.2. Metric <a class="anchor" id="metr"></a>**

We will use the MSE error to optimise our estimator.
$$
MSE(y,\widehat{y}) = \frac{1}{N}\sum_{i=0}^N(y_i-\widehat{y})^2
$$

````python
from sklearn.metrics import make_scorer, mean_squared_error

mse = make_scorer(mean_squared_error, greater_is_better=False)
````

### **3.3. Estimator <a class="anchor" id="estim"></a>**

````python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.dummy import DummyRegressor
from sklearn.ensemble import RandomForestRegressor

reg = DummyRegressor(strategy="mean")
reg2 = RandomForestRegressor(n_estimators=10, max_depth=50, max_features=8)

prep = ColumnTransformer(transformers=[
        ('prep',  make_pipeline(S_imp, StandardScaler()), cols),
        ], remainder='drop')

prep2 = ColumnTransformer(transformers=[
        ('prep',  make_pipeline(K_imp, StandardScaler()), cols),
        ], remainder='drop')

clf = Pipeline(steps=[
    ('prep', prep),('pca',pca),
    ('classifier', reg)])

clf2 = Pipeline(steps=[
    ('prep', prep2),('pca',pca),
    ('classifier', reg2)])

cv = ShuffleSplit(n_splits=5, test_size=0.25)

scores_Xdf = cross_val_score(clf, X, y, cv=cv, scoring=mse)
print("mean: %.2e (+/- %.2e)" % (-scores_Xdf.mean(), scores_Xdf.std()))

scores_Xdf = cross_val_score(clf2, X, y, cv=cv, scoring=mse)
print("mean: %.2e (+/- %.2e)" % (-scores_Xdf.mean(), scores_Xdf.std()))
````

# **4. Submission** <a class="anchor" id="subm"></a>

````python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.ensemble import RandomForestRegressor


def get_estimator():
  K_imp = KNNImputer(missing_values=np.nan, n_neighbors=5, weights="uniform")
  reg = RandomForestRegressor(n_estimators=10, max_depth=50, max_features=8)
  cols = ['P_MHD', 'DAUD', 'PDD', 'PAD', 'PADHD', 'DMSUD', 'PBD',
       'Current health expenditure (% of GDP)',
       'Current health expenditure per capita (current US$)',
       'Out-of-pocket expenditure (% of current health expenditure)',
       'Unemployment, total (% of total labor force) (modeled ILO estimate)',
       'School enrollment, primary (% gross)',
       'School enrollment, secondary (% gross)',
       'School enrollment, tertiary (% gross)', 'ghs',
        'media integrity', 'military expenditure']

  prep = ColumnTransformer(transformers=[
          ('prep',  make_pipeline(K_imp, StandardScaler()), cols),
          ], remainder='drop')

  estimator = Pipeline(steps=[
      ('prep', prep),
      ('classifier', reg)])

  return estimator
````
